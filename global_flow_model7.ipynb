```python
series_partition = Partition.Simple('series', ['1000', '2000', '3000', '4000', '5000', '6000', '7000', '8000', 'cast', 'mixed', 'recycled cans', 'new scrap', 'old scrap'])
palette = {'1000': 'maroon', '2000': 'red', '3000': 'gold', '4000': 'yellowgreen', '5000': 'aqua', '6000': 'navy', '7000': 'purple', '8000': 'magenta', 'cast': 'wheat', 'mixed': 'grey', 'recycled cans': 'darkgreen', 'new scrap': 'peru', 'old scrap': 'saddlebrown'}

sdd = SankeyDefinition(nodes, bundles, ordering, flow_partition=series_partition)

int_range = widgets.IntSlider(value=2020,min=1950,max=2040,step=1)
output2 = widgets.Output()

display(int_range, output2)

def on_value_change(x):
    with output2:
        number = x['new']
        clear_output()
        p = weave(sdd, dataset, measures='{}'.format(number), palette=palette).to_widget(width=1500, height=1050, margins=dict(left=100, right=100))
        display(p)
        
int_range.observe(on_value_change, 'value')
```
